```python
from selenium import webdriver
import csv
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
```

```python
link_data = [];
with open('category.csv', newline='', encoding='utf-8-sig') as csvfile:
    reader = csv.DictReader(csvfile,delimiter=';')
    for row in reader:
        link_data.append(row)
```

```python
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

i = 1
link_product = []
for row in link_data:
    l = row['link']
    driver.get(l)
    for i in range(8):
        driver.execute_script("window.scrollTo(0,"+str(i*710)+")") 
        time.sleep(0.5)
    try:
        element = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, "product-item"))
        )
    except:
        print("Timed out waiting for page to load")
        break
  
    content = driver.page_source #Lấy lại nội dung của website
    soup = BeautifulSoup(content,'html.parser') #Sử dụng BeautifulSoup để xử lý
    
    list_product = soup.findAll('a',attrs={'class':'product-item'})

    for product in list_product:
        quantity = product.find('span',attrs={'class':'quantity'})
        if(quantity):
            quantity = quantity.text.replace("Đã bán","").replace(".","").replace("k","000").replace("+","")
        else:
            quantity = 0
        print(quantity)
        if(int(quantity)>50):
            link = product['href']
            if(link.find("tka.tiki.vn")!=-1):
                link = "https:"+link.replace("https","")
            else:
                link = "https://tiki.vn"+link
            link_product.append([i, link, row['category']])
            i = i + 1

    
    
   
```

```python
print(link_product)
```

```python
import csv
with open('link_san_pham_tiki.csv','wt',newline='\n', encoding='utf-8') as file:
    header = ['id', 'link','category']
    csv_writer = csv.writer(file)
    csv_writer.writerow(header) # write header
    csv_writer.writerows(link_product)
```
